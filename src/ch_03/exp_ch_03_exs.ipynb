```python
"""Adding a vertical barrier"""
def get_horizons(prices, delta=pd.Timedelta(minutes=15)):
    t1 = prices.index.searchsorted(prices.index + delta) # get the index of the timestamps delta minutes ahead
    t1 = t1[t1 < prices.shape[0]] # ignore values beyond the end of the series
    t1 = prices.index[t1] # get the timestamps
    t1 = pd.Series(t1, index=prices.index[:t1.shape[0]]) # create a series with timestamps as index
    return t1
```

```python
"""Now that we have our horizon barriers, we define a function to set upper and 
lower barriers based on the volatility estimates computed earlier:
"""


def get_touches(prices, events, factors=[1, 1]):
    """
    events: pd dataframe with columns
      t1: timestamp of the next horizon
      threshold: unit height of top and bottom barriers
      side: the side of each bet
    factors: multipliers of the threshold to set the height of
             top/bottom barriers
    """
    out = events[['t1']].copy(deep=True)
    if factors[0] > 0:
        thresh_uppr = factors[0] * events['threshold']
    else:
        thresh_uppr = pd.Series(index=events.index)  # no uppr thresh

    if factors[1] > 0:
        thresh_lwr = -factors[1] * events['threshold']
    else:
        thresh_lwr = pd.Series(index=events.index)  # no lwr thresh

    for loc, t1 in events['t1'].items():  # Use items() for Series
        df0 = prices[loc:t1]  # path prices
        df0 = (df0 / prices[loc] - 1) * events.side[loc]  # path returns
        out.loc[loc, 'stop_loss'] = df0[df0 < thresh_lwr[loc]].index.min()  # earliest stop loss
        out.loc[loc, 'take_profit'] = df0[df0 > thresh_uppr[loc]].index.min()  # earliest take profit

    return out
```

```python
"""Finally, we define a function to compute the labels:"""


def get_labels(touches):
    out = touches.copy(deep=True)
    # pandas df.min() ignores NaN values
    first_touch = touches[['stop_loss', 'take_profit']].min(axis=1)
    for loc, t in first_touch.items():  # Use items() instead of iteritems()
        if pd.isnull(t):
            out.loc[loc, 'label'] = 0
        elif t == touches.loc[loc, 'stop_loss']:
            out.loc[loc, 'label'] = -1
        else:
            out.loc[loc, 'label'] = 1
    return out

def get_labels(touches):
    """1 if take profit not null, -1 if stop loss not null, 0 otherwise"""
    out = touches.copy(deep=True)
    out['label']=np.where(~touches.take_profit.isnull(),1,np.where(~touches.stop_loss.isnull(),-1,0))
    return out
```
